```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Carica i dataset
red = pd.read_csv('../winequality-red.csv', sep=';')
white = pd.read_csv('../winequality-white.csv', sep=';')

# Conta le occorrenze di qualità
red_counts = red['quality'].value_counts().sort_index()
white_counts = white['quality'].value_counts().sort_index()

# Gamma completa di qualità
qualities = sorted(set(red_counts.index).union(white_counts.index))

# Allinea i valori ai punteggi di qualità
red_values = [red_counts.get(q, 0) for q in qualities]
white_values = [white_counts.get(q, 0) for q in qualities]

# Parametri per il grafico a barre affiancate
bar_width = 0.4
x = np.arange(len(qualities))

# Colori realistici dei vini
red_color = '#722F37'     # rosso vino scuro
white_color = '#FAFAD2'   # giallo paglierino chiaro

# Plot
plt.figure(figsize=(10, 6))
plt.bar(x - bar_width/2, red_values, width=bar_width, color=red_color, label='Red Wine')
plt.bar(x + bar_width/2, white_values, width=bar_width, color=white_color, label='White Wine')

# Etichette e formattazione
plt.xlabel('Quality')
plt.ylabel('Frequency')
plt.title('Wine quality distribution')
plt.xticks(x, qualities)
plt.legend()
plt.tight_layout()
plt.show()

```
